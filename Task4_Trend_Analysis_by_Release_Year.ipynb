# Task 4: Trend Analysis by Release Year

## Objective

The objective of this task is to analyze how Netflix content production has changed over time using the release year of each title. This analysis helps identify growth trends, peak production years, and historical content patterns.

## import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv("Cleaned_Data/Netflix_Cleaned.csv")

df.head()
```

```python
df.info()
```

```python
df["release_year"].isnull().sum()
```

```python
yearly_content = (
    df["release_year"]
    .value_counts()
    .sort_index()
)

yearly_content.head()
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    yearly_content.index,
    yearly_content.values,
    color="blue",
    linewidth=2
)

plt.title("Netflix Content Released by Year", fontsize=18)

plt.xlabel("Release Year")

plt.ylabel("Number of Titles")

plt.grid(alpha=0.4)

plt.show()
```

## Business Insight

Netflix content production remained relatively low before 2000. The platform experienced rapid growth after 2015, reaching its highest production around 2019. A decline is observed after 2020, indicating changes in production strategy or external factors affecting content releases.

```python
print("Peak Release Year :", yearly_content.idxmax())

print("Highest Number of Titles :", yearly_content.max())
```

```python
top_years = yearly_content.sort_values(
    ascending=False
).head(10)

top_years
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    top_years.index.astype(str),
    top_years.values,
    color="orange"
)

plt.title("Top 10 Release Years")

plt.xlabel("Year")

plt.ylabel("Number of Titles")

plt.xticks(rotation=45)

plt.show()
```

## Business Insight

The highest number of Netflix titles were released in 2019, showing the company's aggressive investment in original content during that period.

```python
fig, ax = plt.subplots(1,2, figsize=(16,6))

# Line Chart
ax[0].plot(
    yearly_content.index,
    yearly_content.values,
    color="blue"
)

ax[0].set_title("Release Year Trend")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Titles")
ax[0].grid(alpha=0.4)

# Top Years
ax[1].bar(
    top_years.index.astype(str),
    top_years.values,
    color="green"
)

ax[1].set_title("Top 10 Release Years")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("Titles")

plt.suptitle(
    "Netflix Release Year Dashboard",
    fontsize=20,
    fontweight="bold"
)

plt.tight_layout()
plt.savefig(
    "Images/Task4_Release_Year_Dashboard.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

## Key Findings

- Netflix content production increased steadily after 2000.
- Rapid growth occurred between 2015 and 2019.
- 2019 recorded the highest number of releases.
- Content production declined after 2020.
- Netflix shifted from limited releases to large-scale content production over time.

## Conclusion

The release year analysis shows that Netflix experienced significant expansion in content production during the late 2010s, particularly between 2015 and 2019. The observed decline after 2020 suggests a shift in production strategy or external market influences. Overall, the trend demonstrates Netflix's rapid growth as a global streaming platform.
