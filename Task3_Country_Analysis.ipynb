# Task 3: Country-wise Netflix Content Analysis

## Data Analysis Using Python Internship

## Objective

The objective of this task is to analyze the geographical distribution of Netflix content and identify the countries contributing the highest number of titles.

This analysis helps understand Netflix's global content distribution and major production regions.

## Libraries Used

- Pandas
- Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Cleaned Dataset

```python
df = pd.read_csv("Cleaned_Data/Netflix_Cleaned.csv")

df.head()
```

## Dataset Preview

```python
df["country"].head(10)
```

## Country-wise Analysis

## Top 10 Countries

```python
country = df["country"].value_counts()

country.head(10)
```

```python
top_countries = country.head(10)

top_countries
```

```python
plt.figure(figsize=(10,6))

bars = plt.barh(
    top_countries.index,
    top_countries.values,
    color="teal",
    edgecolor="black"
)

plt.title("Top 10 Countries with Most Netflix Content", fontsize=16)
plt.xlabel("Number of Titles")
plt.ylabel("Country")

# Add values
for bar in bars:
    plt.text(
        bar.get_width() + 10,
        bar.get_y() + bar.get_height()/2,
        int(bar.get_width()),
        va="center",
        fontsize=10
    )

# Highest value on top
plt.gca().invert_yaxis()

plt.grid(axis="x", linestyle="--", alpha=0.4)

plt.show()
```

## Business Insight

The analysis shows that the **United States** contributes the highest number of Netflix titles, followed by **India** and the **United Kingdom**.

This indicates that Netflix has a strong content presence in North America while also investing significantly in international markets, particularly in India and the United Kingdom.

```python
country = df[df["country"] != "Not Given"]["country"].value_counts()

top_countries = country.head(10)

top_countries
```

```python
df["country"] != "Not Given"
```

## Visualization: Top 10 Countries

```python
plt.figure(figsize=(10,6))

bars = plt.barh(
    top_countries.index,
    top_countries.values,
    color="teal",
    edgecolor="black"
)

plt.title("Top 10 Countries with Most Netflix Content", fontsize=16)
plt.xlabel("Number of Titles")
plt.ylabel("Country")

# Add values
for bar in bars:
    plt.text(
        bar.get_width() + 10,
        bar.get_y() + bar.get_height()/2,
        int(bar.get_width()),
        va="center",
        fontsize=10
    )

# Highest value on top
plt.gca().invert_yaxis()

plt.grid(axis="x", linestyle="--", alpha=0.4)

plt.savefig(
    "Images/Task3_Country_Analysis.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

## Business Insight

The United States contributes the highest number of Netflix titles, followed by India and the United Kingdom.

The analysis also shows that the Top 10 countries contribute **78.01%** of Netflix's entire catalog, indicating that Netflix's content is concentrated among a relatively small number of major production countries.

## Percentage Contribution

The Top 10 countries contribute **78.01%** of the Netflix catalog.

This means that nearly four out of every five Netflix titles originate from just ten countries, highlighting Netflix's strong dependence on key international content-producing markets.

```python
top10_total = top_countries.sum()

overall_total = country.sum()

percentage = (top10_total / overall_total) * 100

print(f"Top 10 Countries contribute {percentage:.2f}% of all Netflix content.")
```
