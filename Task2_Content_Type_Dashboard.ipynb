## Data Analysis Using Python Internship

## Objective

The objective of this task is to analyze the distribution of Movies and TV Shows available on Netflix using the cleaned dataset.

This dashboard helps understand:

- Total Movies
- Total TV Shows
- Percentage distribution
- Overall content composition

## Libraries Used

- Pandas
- Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Cleaned_Data/Netflix_Cleaned.csv")

df.head()
```

## Dataset Preview

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

## Movie and TV Show Count

```python
content = df["type"].value_counts()

content
```

## Visualization 1: Content Distribution

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

bars = plt.bar(
    content.index,
    content.values,
    color=["steelblue", "tomato"],
    edgecolor="black"
)

plt.title("Netflix Content Distribution", fontsize=16)
plt.xlabel("Content Type")
plt.ylabel("Number of Titles")

# Add values above the bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 50,
        int(height),
        ha="center",
        fontsize=11
    )

plt.grid(axis="y", linestyle="--", alpha=0.4)

plt.savefig("Images/Content_Distribution.png", dpi=300, bbox_inches="tight")

plt.show()
```

## Business Insight

The analysis shows that **Movies (6,126 titles)** significantly outnumber **TV Shows (2,664 titles)** on Netflix. This indicates that Netflix's content library is primarily movie-focused while still maintaining a substantial collection of TV shows to serve a wide range of audience preferences.

## Business Insight

The analysis reveals that Movies account for **6,126 titles (69.7%)**, while TV Shows account for **2,664 titles (30.3%)** in the Netflix catalog.

This indicates that Netflix primarily focuses on movies while maintaining a substantial collection of TV shows to provide diverse entertainment for different audience preferences.

## Visualization 2: Percentage Distribution

```python
plt.figure(figsize=(7,7))

plt.pie(
    content.values,
    labels=content.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=["steelblue", "tomato"],
    explode=(0.05, 0)
)

plt.title("Percentage Distribution of Netflix Content")
plt.savefig("Images/Content_Percentage.png",
            dpi=300,
            bbox_inches="tight")
plt.show()
```

## Business Insight

The pie chart shows that **69.7%** of Netflix's content consists of **Movies**, while **30.3%** consists of **TV Shows**. This indicates that Netflix's content strategy is strongly focused on movies, with nearly seven out of every ten titles being films.

## Dashboard

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# --------------------------
# Bar Chart
# --------------------------
bars = axes[0].bar(
    content.index,
    content.values,
    color=["steelblue", "tomato"],
    edgecolor="black"
)

axes[0].set_title("Netflix Content Distribution")
axes[0].set_xlabel("Content Type")
axes[0].set_ylabel("Number of Titles")

for bar in bars:
    height = bar.get_height()
    axes[0].text(
        bar.get_x() + bar.get_width()/2,
        height + 50,
        int(height),
        ha="center"
    )

axes[0].grid(axis="y", linestyle="--", alpha=0.4)

# --------------------------
# Pie Chart
# --------------------------
axes[1].pie(
    content.values,
    labels=content.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=["steelblue", "tomato"],
    explode=(0.05, 0)
)

axes[1].set_title("Percentage Distribution")

# Overall title
plt.suptitle("Netflix Content Type Analysis Dashboard", fontsize=18, fontweight="bold")

plt.tight_layout()
plt.savefig("Images/Content_Dashboard.png",
            dpi=300,
            bbox_inches="tight")
plt.show()
```

## Business Insight

The analysis shows that Movies account for nearly 70% of Netflix's library, while TV Shows contribute around 30%.

This indicates Netflix focuses more on movie production while maintaining a strong TV Show catalog to attract diverse audiences.
