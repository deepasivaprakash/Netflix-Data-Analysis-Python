# Executive Summary

The Netflix dataset contains information about Movies and TV Shows available on the platform. The analysis focused on understanding content distribution, release trends, country contributions, audience ratings, and popular genres.

The goal was to identify patterns that can help explain Netflix's content strategy and support business decisions.

# Key Business Insights

### 1. Content Distribution

- Movies account for approximately **69.7%** of Netflix's catalog.
- TV Shows account for approximately **30.3%**.

---

### 2. Country Analysis

- United States contributes the highest number of titles.
- India ranks second.
- United Kingdom ranks third.
- The Top 10 countries contribute approximately **78.01%** of the Netflix catalog.

---

### 3. Release Trend

- Content production increased rapidly after 2015.
- Peak production occurred around **2018–2019**.
- Releases declined after 2020.

---

### 4. Ratings

- TV-MA is the most common rating.
- TV-14 is the second most common rating.
- Netflix mainly targets mature and teenage audiences.

---

### 5. Genres

- International Movies is the most common genre.
- Dramas and Comedies are also highly represented.
- Netflix maintains a diverse catalog to appeal to a global audience.

# Business Recommendations

Based on the analysis:

1. Continue investing in international content, as International Movies are highly popular.

2. Maintain a balanced catalog of Movies and TV Shows to serve different audience preferences.

3. Expand content production in emerging markets while strengthening existing markets such as India and the United Kingdom.

4. Continue producing mature content while increasing family-friendly content to broaden audience reach.

5. Monitor yearly production trends to maintain a sustainable release strategy.
